## Linear Regression
Linear Regression models the relationship between input variable X and target variable y using:
ŷ = Xθ
Where:
- θ = model parameters (weights)
- X = feature matrix
- ŷ = predicted output

### 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

np.random.seed(42)
```

### 2. Create simple dataset

```python
# Simple dataset (Hours studied vs Score)
X = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([2,4,5,4,5])

# Visualize
plt.scatter(X, y)
plt.title("Dataset")
plt.xlabel("X")
plt.ylabel("y")
plt.show()

# The dataset shows a roughly linear relationship between X and y.
# This makes it suitable for Linear Regression.
```

### 3. Linear Regression(from scratch)

#####  We use the equation:

y = mX + b Hypothesis function for linear regression

#### Initialize Parameters

```python
m = 0   # slope
b = 0   # intercept
learning_rate = 0.01
epochs = 1000
n = len(X)
```

### 4. Gradient decent implementation

```python
for i in range(epochs):
    y_pred = m * X.flatten() + b
    
    # Calculate gradients
    dm = (-2/n) * np.sum(X.flatten() * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)
    
    # Update parameters
    m = m - learning_rate * dm
    b = b - learning_rate * db
    
    # Print loss occasionally
    if i % 100 == 0:
        loss = np.mean((y - y_pred)**2)
        print(f"Epoch {i}, Loss: {loss}")
```

- Gradient Descent minimizes the error step by step.
- Loss decreases over time, showing the model is learning.

### 5. Final model

```python
print("Final slope (m):", m)
print("Final intercept (b):", b)
```

### 6. PLot Regression Line

```python
y_pred = m * X + b

plt.scatter(X, y, label="Actual Data")
plt.plot(X, y_pred, color='red', label="Regression Line")
plt.legend()
plt.title("Linear Regression (From Scratch)")
plt.show()
```

- The regression line fits the data reasonably well.
- Some error exists due to limited dataset size.

### 7. using scikit learn

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)

y_pred_sklearn = model.predict(X)

print("Sklearn slope:", model.coef_)
print("Sklearn intercept:", model.intercept_)
```

### 8. Compared Results

```python
plt.scatter(X, y, label="Actual Data")

# From scratch
plt.plot(X, m*X + b, label="From Scratch")

# Sklearn
plt.plot(X, y_pred_sklearn, linestyle='dashed', label="Sklearn")

plt.legend()
plt.title("Comparison")
plt.show()
```

- Both implementations produce very similar results.
- This confirms our Gradient Descent implementation is correct.

### 10. key learnings
- Linear Regression finds the best-fit line.
- Gradient Descent optimizes parameters (m, b).
- Learning rate controls convergence speed.
- Too high learning rate can cause instability.
